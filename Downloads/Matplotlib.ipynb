# Basic plot

```python
# Basic plot line graph
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,5,14,34,67,98,150])
y = np.array([2,23,34,45,66,78,89])

plt.plot(x,y,linestyle = 'dashed' ) # o:r , o--r
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting the data
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Displaying the plot
plt.show()
```

# grid

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting the data
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.grid()
# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Creating the first subplot (1, 2, 1)
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Subplot 1')

# Creating the second subplot (1, 2, 3)
plt.subplot(1, 2, 2)
plt.plot(y, x)
plt.title('Subplot 2')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

# Bar chart

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [7, 13, 5, 17]

# Create bar chart
plt.bar(categories, values,color = 'r')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Basic Bar Chart')

# Show plot
plt.show()
```

```python
# harizontal bar
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [7, 13, 5, 17]

# Create bar chart
plt.barh(categories, values)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Basic Bar Chart')

# Show plot
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = np.random.randn(1000)

# Create histogram
plt.hist(data, bins=10, edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

# Show plot
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]  # Relative sizes of each section

# Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add title
plt.title('Simple Pie Chart')

# Show plot
plt.show()
```
